```python
from analysis import *
import pandas as pd
from matplotlib import pyplot as plt 
import numpy as np
import re
df = pd.read_csv('../data/all-history-actions.csv')

df['simple_date'] = pd.DatetimeIndex(df['date']).date
```

```python

from cProfile import label
from random import random

from matplotlib.markers import MarkerStyle


df_comm = df.copy(deep=True)
```

```python

# df_comm['committee'] = df_comm[df_comm['action'].str.extract(r"committee on (.*)\W", re.IGNORECASE)]

counts = df_comm.groupby('id').count()

df_comm['committee'] = df_comm['action'].str.extract(r"referred to the committee on ([\w\d\s]*)", re.IGNORECASE)

df_cname = df_comm[~df_comm['committee'].isna()][['id', 'committee', 'simple_date']].head(300)

committee_mention = df_cname.head(50).groupby('id', as_index=True)


plt.figure(figsize=[20, 20])
plt.xticks(rotation=90)


colors = {name.lower() :  np.random.rand(1, 3) for name in df_cname['committee']}

for x in committee_mention.groups:
    grp = committee_mention.get_group(x)
    committees = grp['committee']
    print([name for name in committees])
    plt.scatter(grp['simple_date'], grp['id'], s=100, color=grp['committee'].apply(lambda c : colors[c.lower()]))
    plt.text(50, 50, s="".join(committees))
```

```python
def find_comm_name(x):
  match_comm = re.compile('Committee on (.*)', re.IGNORECASE)
  res = match_comm.search(x)
  if res: 
    return res.groups()[0]
  
df_comm['committee'] = df_comm['action'].apply(find_comm_name)
df_comm
```
